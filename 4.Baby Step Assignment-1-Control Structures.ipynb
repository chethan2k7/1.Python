BMI<16-->Severe Thinness
BMI<20-->Moderate Thinness
BMI<25-->Normal
BMI<35-->Over weight
BMI<40-->Obese
BMI>=40-->Extremely Obese

```python
BMI=int(input("Enter the BMI Index:"))
```

```python
if(BMI<15):
    print("Severe Thinness")
elif(BMI<20):
    print("Moderate Thinness")       
elif(BMI<25):
    print("Normal")   
elif(BMI<35):
    print("Over weight") 
elif(BMI<40):
    print("Obese")    
else:
    print("Extremely Obese")
```
